```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Aman New.txt','r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[AP]M\s-\s(.*)'

messages = re.findall(pattern, data)
messages_list = []
for message in messages:
    messages_list.append(message)

print(messages_list)
```

```python
pattern = '(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2})\s[AP]M\s-\s.*'

dates_and_times = re.findall(pattern, data)
dates_and_times_list = [date for date in dates_and_times]

print(dates_and_times_list)
```

```python
df = pd.DataFrame({'user_message': messages_list, 'message_date': dates_and_times_list })

df['message_date'] = pd.to_datetime(df['message_date'], format= '%m/%d/%y, %H:%M')

df.rename(columns={'messsage_data': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
#separating users and messages
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:#user name
        users.append(entry[1])
        messages.append(entry[2])
    
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages

df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year'] = df['message_date'].dt.year
```
